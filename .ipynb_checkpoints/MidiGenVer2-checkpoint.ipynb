```python
#Create Xlsx file by fixing Pattern Scale with random nots and duration
import random
import numpy

from midiutil.MidiFile import MIDIFile
from openpyxl import Workbook
from pandas import DataFrame

wb = Workbook()
ws = wb.active
ws1 = wb.create_sheet("Pattern")

degrees  = [60, 62, 64, 65, 67, 69, 71, 72] # MIDI note number in C major
degreesm = [60, 62, 63, 65, 67, 68, 70, 72]
length = [0.25,0.5,0.75,1,2]
progession = [0,4,5,3]
channel  = 0
duration = 1   # In beats
volume   = 100 # 0-127, as per the MIDI standard
scaleShift = 2

degreePy = numpy.array(degrees)
degreePym = numpy.array(degreesm)
scale = degreePy+scaleShift
scalem = degreePym+scaleShift

data_pitches = []
data_duration = []

for i in range(20):
    tick = i%4
    tmp_pitches = []
    tmp_duration = []
    leftBeat = 4
    while leftBeat > 0.0:
        print tick
        if tick ==2 :
            selectedScale = numpy.array(scalem)+progession[2]
        else:
            selectedScale = numpy.array(scale)+progession[tick]
        #print leftBeat
        pitch = random.choice(selectedScale)
        duration = random.choice(length)
        while duration > leftBeat:
            duration = random.choice(length)
        tmp_pitches.append(pitch)
        tmp_duration.append(duration)
        leftBeat = leftBeat - duration

    str_pitches = "_".join(str(i) for i in tmp_pitches)
    str_duration = "_".join(str(i) for i in tmp_duration)
    data_pitches.append(str_pitches)
    data_duration.append(str_duration)
    tick = tick+1

df = DataFrame({"Notes":data_pitches,"Duration":data_duration})
df.to_excel("Pattern_20room.xlsx",sheet_name="Pattern",index=False)
print "Generate Pattern Complete"
```

```python
#Get patterns and compose the midi file
import numpy as np
import random

from midiutil.MidiFile import MIDIFile
from openpyxl import load_workbook

wb = load_workbook("Pattern2.xlsx")
ws = wb.active
MyMIDI = MIDIFile(1) # One track, defaults to format 1 (tempo track
                     # automatically created)
pattern_set = range(1,21)
track    = 0
time     = 0   # In beats
tempo    = 120  # In BPM
volume   = 100 # 0-127, as per the MIDI standard
channel  = 0
MyMIDI.addTempo(track,time,tempo)

def main():
    global time
    global track
    global channel
    global volume
    for i in range(8):
        pattern = get_pattern(i+1) #Get Duration and Notes
        pattern_pitches = map(float, pattern[1].split("_"))
        pattern_duration = map(float, pattern[0].split("_"))
        
        for i in range(len(pattern_pitches)):
            pitch = pattern_pitches[i]
            duration = pattern_duration[i]
            MyMIDI.addNote(track, channel, pitch, time, duration, volume)
            time = time+duration
        
    with open("8Room.mid", "wb") as output_file:
        MyMIDI.writeFile(output_file)
        print "MIDI FIle Created"

def get_pattern(n):
    data = []
    for row in ws.iter_rows(min_row = n+1,max_col=2,max_row = n+1):
        for cell in row:
            data.append(cell.value.encode('utf-8')) 
    return data
main()
```
